```python
import twitter_scraper
from twitter_scraper import Profile
from twitter_scraper import get_tweets
import pandas as pd
import re
import bs4
from bs4 import BeautifulSoup
import requests
from time import sleep
import random
import time
import wget as wg
import urllib
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
# In order to get coordinates from a given address, import the following libraries
from geopy.geocoders import Nominatim
import folium
geolocator = Nominatim(user_agent="world_map")
```

```python
users=["AmsterdamNL",
"Berlin_de_News",
"Twiperbole",
"VilleBruxelles",
"koebenhavner",
"DubCityCouncil",
"helsinki",
"lillefrance",
"CamaraLisboa",
"LDN_gov",
"CityLuxembourg",
"MADRID",
"NicosiaMunicipa",
"Oslokommune",
"Paris",
"PrahaEU",
"reykjavik",
"RigasDome",
"Roma",
"Stockholmsstad",
"Vilnius_lt",
"warszawa",
"Stadt_Wien",
"wwwzagrebhr",
"stadtzuerich",
    ]
```

```python
dic={}
dic["Location"]=[]
dic["Username"]=[]
dic["Tweets_count"]=[]
dic["Followers_count"]=[]
dic["Following_count"]=[]
dic["Likes_count"]=[]
dic["Website"]=[]
location=[]
username=[]
tweets_count=[]
followers_count=[]
following_count=[]
likes_count=[]
website=[]
```

```python
for i in users:
    profile = Profile(i)
    dic1=profile.to_dict()
    
    location.append(profile.location)
    username.append(dic1["username"])
    tweets_count.append(dic1["tweets_count"])
    followers_count.append(dic1["followers_count"])
    following_count.append(dic1["following_count"])
    likes_count.append(dic1["likes_count"])
    website.append(dic1["website"])

    dic["Location"]=location
    dic["Username"]=username
    dic["Tweets_count"]=tweets_count
    dic["Followers_count"]=followers_count
    dic["Following_count"]=following_count
    dic["Likes_count"]=likes_count
    dic["Website"]=website
```

```python
# First, we need the right address
address=[]
dic["Lat"]=[]
dic["Lon"]=[]
for i in dic["Location"]:
    try:
        location=geolocator.geocode(i)
        sleep(random.randint(0.0,3.0))   
        address.append(location.address) 
# Second, we add the coordinate of Latitude and Longitude in the dictionary previously created
        
        dic.setdefault("Lat",[]).append(location.latitude)
        dic.setdefault("Lon",[]).append(location.longitude)
    except:
        rifiuti=[]
        rifiuti.append(i)
```

```python
df = pd.DataFrame.from_dict(dic, orient='index')
df=df.transpose()
```

```python
df['Followers_count'] = df['Followers_count'].apply(str)
```
